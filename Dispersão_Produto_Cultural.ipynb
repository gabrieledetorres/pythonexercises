Situação-problema: A empresa Haia Festivais de Música precisa analisar para dois possíveis festivais a combinação das seguintes variáveis: vendas totais e satisfação média relacionadas a produtos de merchandising e exomarketing do evento.

No contexto da gestão de produtos culturais em festivais de música eletrônica, a busca pela excelência na experiência do público e na rentabilidade dos itens comercializados exige métodos analíticos eficazes. Uma das 7 Ferramentas da Qualidade mais relevantes nesse cenário é o Diagrama de Dispersão, utilizado para identificar correlações entre variáveis e orientar decisões estratégicas.

O diagrama de dispersão é uma ferramenta estatística amplamente utilizada para identificar correlações entre variáveis, sendo especialmente útil no setor de vendas para analisar relações entre preço, volume comercializado, campanhas promocionais e desempenho de mercado. Segundo Montgomery e Runger (2012), “os diagramas de dispersão são essenciais para a identificação visual de padrões entre variáveis contínuas, permitindo insights sobre tendências e possíveis relações causais que impactam diretamente decisões estratégicas em ambientes comerciais”. Essa abordagem facilita a tomada de decisões baseadas em dados e contribui para o aprimoramento da qualidade nos processos de vendas.

### Os dados para esta análise foram obtidos através de gestão de conhecimento de festivais realizados anteriormente pela Haia Festivais de Música. [Case fictício]
Foram coletados dados de satisfação (escala de 0 a 10) e volume de vendas de produtos como camisetas, bonés, acessórios LED, entre outros.
Os dados foram plotados em um diagrama de dispersão, com cada ponto representando um produto em um dos festivais.
A análise visual permitiu identificar padrões de correlação entre as variáveis.



RESULTADOS E DISCUSSÕES

Portanto, foi realizado o mapeamento de produtos e mensuração de satisfação divididos em duas cores para prever o cenário nos dois festivais conforme o Diagrama de Dispersão na Figura 01


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

sns.set(style="whitegrid")

output_dir = "/mnt/data"
os.makedirs(output_dir, exist_ok=True)

np.random.seed(42)
produtos = ['Camisetas', 'Bonés', 'Pulseiras', 'Óculos', 'Posters', 'Bebidas Temáticas', 'Acessórios LED']

festival_a = pd.DataFrame({
    'Produto': produtos,
    'Satisfacao': np.random.uniform(6, 10, len(produtos)),
    'Vendas': np.random.randint(100, 1000, len(produtos)),
    'Festival': 'Festival A'
})

festival_b = pd.DataFrame({
    'Produto': produtos,
    'Satisfacao': np.random.uniform(5, 9, len(produtos)),
    'Vendas': np.random.randint(80, 950, len(produtos)),
    'Festival': 'Festival B'
})


dados = pd.concat([festival_a, festival_b], ignore_index=True)

plt.figure(figsize=(12, 8))
sns.scatterplot(data=dados, x='Satisfacao', y='Vendas', hue='Festival', style='Produto', palette='plasma', s=150)
plt.title('Análise de Produtos Culturais em Festivais de Música Eletrônica', fontsize=16)
plt.xlabel('Satisfação Média (0-10)', fontsize=12)
plt.ylabel('Vendas Totais', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plot_path = os.path.join(output_dir, "dispersao_produtos_festivais.png")
plt.show()


print("Figura 01: Gráfico de Dispersão - Análise de venda dos produtos para dois festivais")
```

Na análise gerada, observamos que para os festivais obtemos os seguintes parâmetros de vendas e satisfação:

Festival A

Menor satisfação: Acessórios de LED
Menor previsão de vendas: Camisetas

Maior satisfação: Bonés
Maior previsão de vendas: Óculos

Melhor escolha combinada dos dois critérios: Óculos



Festival B

Menor satisfação: Óculos
Menor previsão de vendas: Bebidas Temáticas

Maior satisfação: Pôsters
Maior previsão de vendas: Pulseiras

Melhor escolha combinada dos dois critérios: Pulseiras


CONCLUSÃO

Produtos como Camisetas e Acessórios LED apresentam alta satisfação e boas vendas em ambos os festivais — indicam forte aceitação e potencial de investimento.

Posters e Bonés têm desempenho mais modesto, sugerindo necessidade de redesign ou reposicionamento.

O cruzamento entre satisfação e vendas permite identificar produtos com boa percepção, mas baixa conversão — ideal para ações de marketing ou melhoria de experiência.

Porém, para vender óculos simultaneamente nos dois festivais, sugere-se que para o Festival B, observe-se a cadeia de suprimentos e atendimento de qualidade. Assim como para a venda de camisetas no Festival A, precisa-se de uma reformulação na estratégia de vendas - conforme índice de vendas mediano apresentado no gráfico.
